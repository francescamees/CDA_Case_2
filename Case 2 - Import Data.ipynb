# Load the Data Set

```python
# File path
file_name = 'data\\HR_data.csv'
data = pd.read_csv(file_name)
```

```python
print("Shape of the Data: ",np.shape(data))
# Observations (rows) with at least one NA
print("Number of Observations with NA: ",pd.isnull(data.transpose()).any().sum())
# Parameters (columns) with at least one NA
print("Number of Parameters with NA: ",pd.isnull(data).any().sum())
```

```python
# Positions of NA values (as boolean mask)
na_positions = data.isnull()

# Get exact (row, column) index pairs where NA occurs
na_indices = np.argwhere(na_positions.values)
na_labeled = [(data.index[row], data.columns[col]) for row, col in na_indices]

print("\nExact positions of NA values (row index, column name):")
for pos in na_labeled:
    print(pos)
```

```python
data['Cohort'] = data['Cohort'].str.replace('D1_', '').astype(int)
data['Round'] = data['Round'].str.replace('round_', '').astype(int)
data['Phase'] = data['Phase'].str.replace('phase', '').astype(int)
```

```python
# Replace missing values with median
data = data.fillna(data.median())
```

```python
data_bio = data[data.columns[1:52]]
data_stats = data[['Cohort','Individual','Round','Phase','Puzzler']]
data_emo = data[ [data.columns[-12]] + list(data.columns[-10:])]
```

```python
from sklearn.preprocessing import StandardScaler
# compute required values
scaler = StandardScaler()
bio_scaler = scaler.fit(data_bio)
data_bio = pd.DataFrame(bio_scaler.transform(data_bio), columns=data.columns[1:52])

data_stats = data_stats.astype('int8')
data_emo['Frustrated'] *= 1/2 
data_emo = data_emo.astype('float16')
```

```python
# Concatenate scaled info, signals and emotions
data_processed = pd.concat([data_stats, data_bio, data_emo], axis=1)
data_processed.head()
data_processed.info()

# Export data as csv
data_processed.to_csv('data\\data_pre_processed.csv', index=False)
```
